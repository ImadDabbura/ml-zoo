```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from scipy.io import loadmat

%matplotlib inline
sns.set_context("notebook")
plt.style.use("fivethirtyeight")
```

```python
data = loadmat("MachineLearning_Coursera/Machine-Learning/data/ex7data2.mat")
X = data["X"]
X.shape
```

```python
km = KMeans(n_clusters=3, init="k-means++", max_iter=100, tol=1e-4)
idx = km.fit_predict(X)
centroids = km.cluster_centers_
# plot the data colored by the clusters each example belongs to
fig, ax = plt.subplots(figsize = (10, 8))
plt.scatter(X[idx == 0, 0], X[idx == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[idx == 1, 0], X[idx == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(X[idx == 2, 0], X[idx == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', s = 80, c = 'k', label = 'Cluster Centroid')
plt.legend()
plt.title('K-Means Clustering Results with K=3');
```

```python
sse = []
for i in range(1, 21):
    km = KMeans(n_clusters=i, init="k-means++", max_iter=100, tol=1e-4)
    km.fit(X)
    sse.append(km.inertia_)

plt.plot(range(1, 21), sse, marker="o")
```

```python
hc = AgglomerativeClustering(n_clusters=3, linkage="complete")
idx = hc.fit_predict(X)
# plot the data colored by the clusters each example belongs to
fig, ax = plt.subplots(figsize = (10, 8))
plt.scatter(X[idx == 0, 0], X[idx == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[idx == 1, 0], X[idx == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(X[idx == 2, 0], X[idx == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.legend()
plt.title('Agglomerative Clustering Results with K=3');
```

```python
dbc = DBSCAN(eps=0.2, min_samples=12)
idx = dbc.fit_predict(X)
# plot the data colored by the clusters each example belongs to
fig, ax = plt.subplots(figsize = (10, 8))
plt.scatter(X[idx == 0, 0], X[idx == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[idx == 1, 0], X[idx == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(X[idx == -1, 0], X[idx == -1, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.legend()
plt.title('Agglomerative Clustering Results with K=3');
```
